```python
df_country=pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df.columns
```

```python
final_df=pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df.head(2)
```

```python
# To check data types
final_df.dtypes
```

```python
final_df.Country.value_counts()
```

```python
country_names=final_df.Country.value_counts().index
country_names
```

```python
country_val=final_df.Country.value_counts().values
country_val
```

```python
#Pie Chart for top 3 countries that use zomota
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%') 
```

```python
## Observation : Zomoto maximun records or tranction are from India after that USA and then United Kingdom
```

```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
ratings
```

```python
'''Observation 
1. When Ratings is between 4.5 to 4.9 --> Excellent
2. When Ratings is between 4.0 to 3.4 --> Very Good
3. When Ratings is between 3.5 to 3.9 --> Good
4. When Ratings is between 3.0 to 3.4--> Average
5. When Ratings is between 2.5 to 2.9 --> Average
6. When Ratings is between 2.0 to 2.4 --> Poor
'''
```

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x="Aggregate rating",y="Rating Count",data= ratings)
```

```python
sns.barplot(x="Aggregate rating",y="Rating Count",hue='Rating color',data= ratings,palette=['blue','red','orange','yellow','green','green'])
```

```python
'''Observation:
1. Not Rated count is very high
2. Maximum no. of rating are between 2.5 to 3.4'''
```

```python
## Count plot
sns.countplot(x='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','green'])
```
